```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Create 6 data points
data_points = np.array([-4, -1, 0, 1, 4, 6])

# Generate small bell curves around each data point
x = np.linspace(-8, 10, 1000)
curves = []
for point in data_points:
    curve = 0.05*norm.pdf(x, loc=point, scale=0.5)
    curves.append(curve)
    plt.plot(x, curve, 'r--', lw=2, label='Gaussian' if point == data_points[0] else "")

# Combine the curves into a single dataset
data = np.hstack([np.random.normal(loc=point, scale=0.5, size=100) for point in data_points])

# Plot the KDE of the combined data
sns.kdeplot(data, color='blue', linewidth=2, label='KDE')
plt.vlines(data_points, ymin=0, ymax=0.005, color='black', lw=2, label='Data Point')
# Show the plot
plt.xlabel('X-axis')
plt.ylabel('Density function')
plt.legend()
plt.show()

```
